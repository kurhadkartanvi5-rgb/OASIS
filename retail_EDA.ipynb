```python
sns.countplot(x="customer_gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x="customer_segment", data=df)
plt.title("Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="category", data=df)
plt.title("Product Categories")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("cleaned_retail_sales.csv", index=False)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["sales_amount"], bins=20)
plt.title("Distribution of Total Sales")
plt.xlabel("Total Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["sales_amount"])
plt.title("Box Plot of Sales Amount")
plt.ylabel("Sales Amount")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["customer_gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["customer_segment"].value_counts().plot(kind="bar")
plt.title("Customer Segment Distribution")
plt.xlabel("Customer Segment")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["category"].value_counts().plot(kind="bar")
plt.title("Product Category Distribution")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["payment_method"].value_counts().plot(kind="bar")
plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include="number").corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

# Conclusion

The Exploratory Data Analysis (EDA) of the Retail Sales dataset was successfully completed. The dataset was checked for missing values, duplicates, and data types.
